```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import glob
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df.head()
```

# UNIVARIATE PLOT

```python
sns.histplot(df['not_distracted'])
```

```python
sns.displot(df['not_distracted'], kde=True)
sns.set(rc={'figure.figsize':(8,4)})
```

```python
sns.kdeplot(df['not_distracted'])
sns.set(rc={'figure.figsize':(8,4)})
```

```python
df['not_distracted'].plot()
plt.figure.figsize=(8,4)
```

```python
df['not_distracted'].plot.hist()
```

# BIVARIATE PLOT

```python
sns.jointplot(x='speeding', y='alcohol', data=df)
sns.set(rc={'figure.figsize':(8,4)})
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
sns.set(rc={'figure.figsize':(8,4)})
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='kde')
sns.set(rc={'figure.figsize':(8,4)})
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='hex')
sns.set(rc={'figure.figsize':(8,4)})
```

```python
sns.barplot(x='total', y='abbrev', data=df)
sns.set(rc={'figure.figsize':(30,15)})
```
